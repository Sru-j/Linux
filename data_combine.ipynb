```python
# Find duplicates based only on 'Name' column
duplicates_subset = df_center[df_center.duplicated(subset=['customer_name'], keep=False)]

print("\nDuplicate Rows (based on Name):")
print(duplicates_subset)
```

```python
# print the number of duplicate rows

count_duplicates = df_center.duplicated().sum()
print(f"Number of duplicate rows: {count_duplicates}")
```

```python
# print the duplicates

duplicates_only = df_center[df_center.duplicated(keep=False)]

print("Duplicate Rows Only:")
print(duplicates_only)
```

```python
#print the duplicate values on customer

duplicates_only = df_center[df_center.duplicated(keep=False)]

print("Duplicate Rows Only:")
print(duplicates_only['customer_id'])
```

```python
# percentage of duplicates
total = len(df_center)
percentage = (count_duplicates/total)*100
print(f"Percentage of duplicates: {percentage:.2f}%")
```

```python
#delete duplicates

df_cleaned = df_center.drop_duplicates()
df_cleaned.shape
```
